```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
cityData = pd.read_csv(city_data_to_load, encoding="utf-8")
rideData = pd.read_csv(ride_data_to_load, encoding="utf-8")

# Combine the data into a single dataset
combinedDf = pd.merge(rideData, cityData, on="city")



# Display the data table for preview
combinedDf.sort_values(by="city", ascending=True) 
combinedDf.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
#create df with average fair and number of rides per city by ride_id
rideCount_avgFare = pd.DataFrame({
    "number of rides" : combinedDf.groupby("city")["ride_id"].count(),
    "average fare" : combinedDf.groupby("city")["fare"].mean()
})
#add in type with merge by city
rideCount_avgFare = pd.merge(rideCount_avgFare, cityData, on="city")
rideCount_avgFare


#groupedCombinedDF
urbanDataDf = rideCount_avgFare.loc[rideCount_avgFare["type"] == "Urban"]
suburbanDataDf = rideCount_avgFare.loc[rideCount_avgFare["type"] == "Suburban"]
ruralDataDf = rideCount_avgFare.loc[rideCount_avgFare["type"] == "Rural"]
urbanDataDf
suburbanDataDf
ruralDataDf

# Build the scatter plots for each city types
plt.scatter(
    urbanDataDf["number of rides"], urbanDataDf["average fare"], s=urbanDataDf["driver_count"]*10, 
    color = "coral", alpha=0.8, edgecolors="black", linewidth=1
)
plt.scatter(
    suburbanDataDf["number of rides"], suburbanDataDf["average fare"], s=suburbanDataDf["driver_count"]*10, 
    color = "c", alpha=0.8, edgecolors="black", linewidth=1
)
plt.scatter(
    ruralDataDf["number of rides"], ruralDataDf["average fare"], s=ruralDataDf["driver_count"]*10, 
    color = "y", alpha=0.8, edgecolors="black", linewidth=1
)

# Incorporate the other graph properties
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare($)")
plt.title("Pyber Ride Sharing Data (2016)")
plt.xticks(np.arange(0, 41, 5))
plt.yticks(np.arange(20, 41, 5))
plt.grid(b=None, which='major', axis='both')

# Create a legend
lgnd = plt.legend(labels=["Urban", "Subrban", "Rural"], loc="best", markerscale=.5)

# Incorporate a text label regarding circle size
#fig = plt.figure()
#plt.annotate('local max', xy=(0, 0), xytext=(0, 0), size=20)
plt.text(45, 35, "Note:\nCircle size correlates with driver count per city.", fontdict=None, withdash=False)

lgnd.legendHandles[0]._sizes = [30]
lgnd.legendHandles[1]._sizes = [30]
lgnd.legendHandles[2]._sizes = [30]


# Save Figure
plt.savefig('Pyber Ride Sharing Data (2016).png')
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
FareTotalDf = pd.DataFrame({"Fare Total" : combinedDf.groupby("type")["fare"].sum()})
FareTotalDf["Fare $ Percentage"] = FareTotalDf["Fare Total"]/FareTotalDf["Fare Total"].sum()*100
FareTotalDf

# Build Pie Chart
pieColors = ("gold", "skyblue", "lightcoral" )

FareTotalDf.plot(y = "Fare $ Percentage", 
                 kind = "pie", 
                 explode = (0,0,0.1), 
                 shadow = True,
                 startangle = 150, 
                 labels = FareTotalDf.index,
                 legend = False,
                 autopct = '%1.1f%%',
                 colors = pieColors,
                 title = "% of Total Fares by City Type",
                 label = ""
                )
# Save Figure
plt.savefig('% of Total Fares by City Type.png')
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
FareCountsDf = pd.DataFrame({"Fare Counts" : combinedDf.groupby("type")["ride_id"].count()})
FareCountsDf["Fare Count Percentage"] = FareCountsDf["Fare Counts"]/FareCountsDf["Fare Counts"].sum()*100
FareCountsDf

# Build Pie Chart
pieColors = ("gold", "skyblue", "lightcoral" )

FareCountsDf.plot(y = "Fare Count Percentage", 
                 kind = "pie", 
                 explode = (0,0,0.1), 
                 shadow = True,
                 startangle = 150, 
                 labels = FareTotalDf.index,
                 legend = False,
                 autopct = '%1.1f%%',
                 colors = pieColors,
                 title = "% of Total Rides by City Type",
                 label = ""
                )

# Save Figure
plt.savefig('% of Total Rides by City Type.png')
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Driver Percents
DriverCountsDf = pd.DataFrame({"Total Drivers" : cityData.groupby("type")["driver_count"].sum()})
DriverCountsDf["Total Drivers Percentage"] = DriverCountsDf["Total Drivers"]/DriverCountsDf["Total Drivers"].sum()*100
DriverCountsDf

# Build Pie Chart
pieColors = ("gold", "skyblue", "lightcoral" )

DriverCountsDf.plot(y = "Total Drivers Percentage", 
                 kind = "pie", 
                 explode = (0,0,0.1), 
                 shadow = True,
                 startangle = 150, 
                 labels = FareTotalDf.index,
                 legend = False,
                 autopct = '%1.1f%%',
                 colors = pieColors,
                 title = "% of Total Rides by City Type",
                 label = ""
                )

# Save Figure
plt.savefig('% of Total Drivers by City Type.png')
plt.show()
```
